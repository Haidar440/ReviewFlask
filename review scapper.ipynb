```python
req = Request(
    product_link,
    headers={
        "User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                      "AppleWebKit/537.36 (KHTML, like Gecko) "
                      "Chrome/124.0.0.0 Safari/537.36"
    }
)
prodRes = urlopen(req)
prodRes = prodRes.read()
```

```python
prod_html = bs(prodRes, "html.parser")
prod_html
```

```python
commentboxes = prod_html.find_all('div', {'class': "col EPCmJX"})
```

```python
len(commentboxes)
```

```python
print("{:<25} | {:<8} | {:<30} | {:<100}".format('Reviewer Name', 'Rating', 'Heading', 'Comment'))
print("-"*170)
for commentbox in commentboxes:
    try:
        name = commentbox.find_all('p', {'class': '_2NsDsF AwS1CA'})[0].text
    except:
        name = ''
    try:
        rating = commentbox.div.div.text
    except:
        rating = ''
    try:
        heading = commentbox.div.p.text
    except:
        heading = ''
    try:
        comment = commentbox.find_all('div',{'class':''})[0].text
    except:
        comment = ''
    try:
        comtag = commentbox.find_all('div',{'class':'ZmyHeo'})
        extra = comtag[0].div.div.text if comtag else ''
    except:
        extra = ''
    full_comment = comment + (' -- ' + extra if extra else '')
    print("{:<25} | {:<8} | {:<30} | {:<100}".format(name, rating, heading, full_comment))
```
